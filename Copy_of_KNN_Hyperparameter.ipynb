```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
k_range = range(1, 31)
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10)  # 10-fold x-validation
    k_scores.append(scores.mean())

# Plot
plt.figure(figsize=(10, 5))
plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
k_range_zoomed = range(11, 16)
k_scores_zoomed = []

for k in k_range_zoomed:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=10)
    k_scores_zoomed.append(scores.mean())

# Plot
plt.figure(figsize=(10, 5))
plt.plot(k_range_zoomed, k_scores_zoomed)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
param_grid = {'n_neighbors': np.arange(1, 31)}
knn = KNeighborsClassifier()
knn_gscv = GridSearchCV(knn, param_grid, cv=10)
knn_gscv.fit(X, y)

print("Best K value:", knn_gscv.best_params_)
print("Best Cross-Validated Accuracy:", knn_gscv.best_score_)
```
