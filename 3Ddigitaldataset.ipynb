```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
```

```python
# load digits dataset
digits = load_digits()
X = digits.data
y = digits.target
```

```python
# TSNE estimator reduces data to three dimensions
# TSNE estimator
tsne = TSNE(n_components=3, random_state=42)
# Reduce datasets dimensions
reduced_data = tsne.fit_transform(X)
```

```python
# Create 3d scatterplot
figure = plt.figure(figsize=(9, 9))
axes = figure.add_subplot(111, projection='3d')
# Plot pts
dots = axes.scatter(xs=reduced_data[:, 0],
                    ys=-reduced_data[:, 1],
                    zs=reduced_data[:, 2],
                    c=digits.target,
                    cmap=plt.cm.get_cmap('nipy_spectral_r', 10))
# Map colors to digits
color_bar = plt.colorbar(dots, ax=axes)
color_bar.set_label('Digit Label')

plt.show()
```
